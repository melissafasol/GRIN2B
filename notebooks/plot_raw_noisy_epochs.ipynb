```python
start_time_df_list_br_1 = []
start_time_df_list_br_2 = []
quantifying_indices = []
for animal in br_animal_IDs:
    os.chdir(path_to_brainstate_files)
    br_1 = pd.read_pickle(str(animal) + '_BL1.pkl')
    br_2 = pd.read_pickle(str(animal) + '_BL2.pkl')
    noisy_epochs_1 = noisy_idx_br_1.loc[noisy_idx_br_1['Animal_ID'] == int(animal)]
    noisy_epochs_2 = noisy_idx_br_2.loc[noisy_idx_br_2['Animal_ID'] == int(animal)]
    #CHECK THAT THERE ARE NO DUPLICATE INDICES 
    indices_br_1 = list(set(noisy_epochs_1['epoch_idx'].tolist()))
    indices_br_2 = list(set(noisy_epochs_2['epoch_idx'].tolist()))
    quantifying_noisy_indices = {'Animal_ID': [animal], 'number_noisy_epochs' : [len(indices_br_1) + len(indices_br_2)]}
    quantifying_indices.append(pd.DataFrame(data = quantifying_noisy_indices))
    br_state_noisy_1 = br_1.loc[br_1['brainstate'] == brain_state_number]
    br_state_noisy_2 = br_2.loc[br_2['brainstate'] == brain_state_number]
    br_1_start_times = []
    br_2_start_times = []
    if len(indices_br_1) > 0:
        for idx_1 in indices_br_1:
            start_epoch = br_state_noisy_1.iloc[idx_1]['start_epoch']
            br_1_dict = {'animal_id': [animal], 'start_time_seconds': [start_epoch]}
            br_1_start_times.append(pd.DataFrame(data = br_1_dict))
    else:
        pass 
    if len(indices_br_2) > 0:
        for idx_2 in indices_br_2:
            start_epoch = br_state_noisy_2.iloc[idx_2]['start_epoch']
            br_2_dict = {'animal_id': [animal], 'start_time_seconds': [start_epoch]}
            br_2_start_times.append(pd.DataFrame(data = br_2_dict))
    else:
        pass
    
    if len(br_1_start_times) > 0:
        start_time_df_list_br_1.append(pd.concat(br_1_start_times))
    else:
        pass
    
    if len(br_2_start_times) > 0:
        start_time_df_list_br_2.append(pd.concat(br_2_start_times))
    else:
        pass
    
```

```python
br_1_df = pd.concat(start_time_df_list_br_1)
br_2_df = pd.concat(start_time_df_list_br_2)
idx_quantify = pd.concat(quantifying_indices)
```
